### Feature selection

Let us use hypothesis testing for feature selection process

***Statistical testing of continuous feature with target feature***

```python
for i in num_cols:
    x = np.array(df[df.Churn=='No'][i])
    y = np.array(df[df.Churn=='Yes'][i])
    
    t, p_value  = stats.ttest_ind(x,y, axis = 0,equal_var=False) 
    
    print('{} P_Value:{}'.format('\033[1m',p_value))
    if p_value < 0.05:  # Setting our significance level at 5%
        print('{} Rejecting Null Hypothesis. \n There is significant difference in {} Feature for different category of target variable'.format('\033[1m',i))
    else:
        print('{} Fail to Reject Null Hypothesis.\n There is no significant difference in {} Feature for different category of target variable'.format('\033[1m',i))
    print('\n')
```

The features "tenure","MonthlyCharges","TotalCharges" are required for our model building

***Statistical testing of categorical feature with target feature***

```python
for i in obj_cols:
    crosstab=pd.crosstab(df["Churn"],df[i])
    
    chi,p_value,dof,expeced = stats.chi2_contingency(crosstab)
    
    if p_value < 0.05:  # Setting our significance level at 5%
        print('{} Rejecting Null Hypothesis. \n There is significant difference in {} Feature for different category of target variable'.format('\033[1m',i))
    else:
        print('{} Fail to Reject Null Hypothesis.\n There is no significant difference in {} Feature for different category of target variable'.format('\033[1m',i))
    print(crosstab)    
    print('\n')
```

We can see gender,PhoneService does not have effect on target variable. So dropping these columns before building model

```python
df.drop(['gender','PhoneService'],axis=1,inplace=True)
```

## 5. Data preprocessing

### Encoding target variable

```python
df['Churn']=df['Churn'].replace({"Yes": 1,"No": 0})
```

### Outlier analysis

During Univariate analysis we have seen that there are no outliers in the dataset, so there is no need for analysing the outliers

### Checking for Target imbalance

```python
f,axes=plt.subplots(1,2,figsize=(17,7))
df['Churn'].value_counts().plot.pie(autopct='%1.1f%%',ax=axes[0],shadow=True)
sns.countplot('Churn',data=df,ax=axes[1],order=[0,1])
axes[0].set_title('Churn Variable Pie Chart')
axes[1].set_title('Churn Variable Bar Graph')
plt.show()
```

We can see that the Target feature is highly imbalanced.we need to balance the dataset before building the model.

### Segregatiing predictors vs target features

```python
X=df.drop(columns="Churn")
y=df["Churn"]
```

Since there are no Ordinal categorical features we can go for One-Hot encoding

**One hot encoding**

```python
X=pd.get_dummies(X,drop_first=True)
```

**Scaling the continuous features**

```python
scaler = StandardScaler()
temp=X[num_cols] #Temporary dataframe
temp[num_cols]= scaler.fit_transform(temp)
```

Applying the scaled data to the dataframe X

```python
X[num_cols]=temp
```

Saving the cleaned dataset as CSV file for future use

```python
new_df=X.join(y,how='inner')
new_df.to_csv("clean_dataset")
```

### Train-Test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10,stratify = y)
```

## 6. Model training,testing and tuning

### Decision Tree

```python
dTree = DecisionTreeClassifier(criterion = 'gini', random_state=1)
dTree.fit(X_train, y_train)
```

```python
print("Decision Tree Training data score: ",dTree.score(X_train, y_train))
print("Decision Tree Testing data score: ",dTree.score(X_test, y_test))
```

Our model is Overfitting

***Reducing over fitting (Regularization)***

```python
dTreeR = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3, random_state=1)
dTreeR.fit(X_train, y_train)
print("Decision Tree Training data score: ",dTreeR.score(X_train, y_train))
print("Decision Tree Testing data score: ",dTreeR.score(X_test, y_test))
```

```python
dTreeR = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, random_state=1)
dTreeR.fit(X_train, y_train)
print("Decision Tree Training data score: ",dTreeR.score(X_train, y_train))
print("Decision Tree Testing data score: ",dTreeR.score(X_test, y_test))
```

***Insight from the above models***

criterion='gini' yields better score than criterion = 'entropy' .So let us choose the model with criterion = 'gini'

***Important features in tree building***

```python
pd.DataFrame(dTreeR.feature_importances_, columns = ["Imp"], index = X_train.columns)
```

***Insights***

tenure,InternetService_Fiber optic,Contract_Two year are the features that impact Churn 

***Confusion Matrix for Decision Tree***

```python
print("Decision Tree Testing data score: ",dTreeR.score(X_test , y_test))
y_predict = dTreeR.predict(X_test)

cm = confusion_matrix(y_test, y_predict, labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No","Yes"]],
                  columns = [i for i in ["No","Yes"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
plt.show()
```

***Classification report for Decision Tree Classifier***

```python
print("classification  Matrix:\n",classification_report(y_test,y_predict))
```

### Ensemble Learning - Bagging

```python
bgcl = BaggingClassifier(base_estimator=dTree, n_estimators=10,random_state=1)

bgcl = bgcl.fit(X_train, y_train)

print("Ensemble learning(Bagging) Training data score: ",bgcl.score(X_train,y_train))
print("Ensemble learning(Bagging) Testing data score: ",bgcl.score(X_test,y_test))
```

Our model is overfitting, we need to do hyperparameter tuning

***Confusion Matrix for Bagging Classifier***

```python
y_predict = bgcl.predict(X_test)

print(bgcl.score(X_test , y_test))

cm=confusion_matrix(y_test, y_predict,labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No","Yes"]],
                  columns = [i for i in ["No","Yes"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

***Classification report for Bagging Classifier***

```python
print("classification  Matrix:\n",classification_report(y_test,y_predict))
```

### Ensemble Learning - AdaBoosting

```python
abcl = AdaBoostClassifier(n_estimators=10, random_state=1)
abcl = abcl.fit(X_train, y_train)

print("Ensemble learning(AdaBoosting) Training data score: ",abcl.score(X_train,y_train))
print("Ensemble learning(AdaBoosting) Testing data score: ",abcl.score(X_test,y_test))
```

***Confusion Matrix for AdaptiveBoosting Classifier***

```python
y_predict = abcl.predict(X_test)
print(abcl.score(X_test , y_test))

cm=confusion_matrix(y_test, y_predict,labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No","Yes"]],
                  columns = [i for i in ["No","Yes"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

***Classification report for AdaptiveBoosting Classifier***

```python
print("classification  Matrix:\n",classification_report(y_test,y_predict))
```

### Ensemble Learning - GradientBoost

```python
gbcl = GradientBoostingClassifier(n_estimators = 50,random_state=1)
gbcl = gbcl.fit(X_train, y_train)

print("Ensemble learning(GradientBoosting) Training data score: ",gbcl.score(X_train,y_train))
print("Ensemble learning(GradientBoosting) Testing data score: ",gbcl.score(X_test,y_test))
```

***Confusion Matrix for GradientBoosting Classifier***

### Ensemble RandomForest Classifier

```python
rfcl = RandomForestClassifier(n_estimators = 50, random_state=1,max_features=12)
rfcl = rfcl.fit(X_train, y_train)

print("Ensemble learning(RandomForest) Training data score: ",rfcl.score(X_train,y_train))
print("Ensemble learning(RandomForest) Testing data score: ",rfcl.score(X_test,y_test))
```

### Final model

***We can see that from all the above models Gradient boosting is performing well, it has higher accuracy and f1 score, So we will choose GradientBoosting as our final model.***
